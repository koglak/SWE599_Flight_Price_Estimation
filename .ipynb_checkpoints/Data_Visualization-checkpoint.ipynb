```python
sns.countplot(x=df["arrival_city"], order = df['arrival_city'].value_counts().index, color='#5975a4', saturation=1).set_title('Distribution of Flight Count of Cities', fontsize=20)
plt.xticks(rotation=70, fontsize=12)
plt.xlabel('Arrival Cities',fontsize=15)
plt.ylabel('Number of Flights',fontsize=15)
```

```python

f,ax=plt.subplots(1,2,figsize=(22,8))

ax[0] = df["company"].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[0], fontsize=13, colors = ['#34495E', '#5D6D7E', '#AEB6BF'])
ax[0].set_title("Distribution of Airlines", fontsize=20)
ax[0].legend(bbox_to_anchor=(1, 1), fontsize=12)

ax[1] = sns.countplot(x="company", data=df, palette=['#34495E', '#5D6D7E', '#AEB6BF'])
ax[1].set_title("Frequency distribution of airlines", fontsize=20)
plt.xlabel('Airlines',fontsize=15)
plt.ylabel('Number of Flights',fontsize=15)

plt.show()
```

```python

for city in arrival_cities:
    df_filtered_by_city=df[df["arrival_city"]==city]
    x = pd.Series(df_filtered_by_city['price_usd'], name="Price($)")
    ax = sns.kdeplot(x, color='#34495E', fill=True)
    ax.set_title("Price($) Distribution for {}".format(city), fontsize=14)
    plt.show()
```

```python
plt.figure(figsize=(20,10))
corr = df.corr()
sns.heatmap(corr,annot=True,cmap='Blues')
plt.show()
```

```python
sns.boxplot(y = "price_usd", x = "company", data = df.sort_values("price_usd", ascending = False), palette=['#34495E', '#5D6D7E','#AEB6BF'])
plt.show()
```

```python
sns.boxplot(y = "price_usd", x = "departure_airport", data = df.sort_values("price_usd", ascending = False),palette=['#34495E', '#AEB6BF'])

plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,9)})


ax=sns.boxplot(x="arrival_city", y="price_usd", data=df, palette=['#38424d','#34495E','#566573', '#5D6D7E', '#85929E', '#AEB6BF','#EBEDEF'])
ax.set_title('Boxplot for Arrival City vs Price ($)', fontsize=20)
ax.set_xlabel('Arrival City',fontsize=14)
ax.set_ylabel('Price ($)',fontsize=14)
```

```python
sns.boxplot(y = "price_usd", x = "departure_week_day", data = df, palette=['#38424d','#34495E','#566573', '#5D6D7E', '#85929E', '#AEB6BF','#EBEDEF'])
plt.show()
```

```python
sns.catplot(x='company',y='price_usd',hue='departure_airport',kind='violin',data=df)
```

```python

f,ax=plt.subplots(1,2,figsize=(22,8))

ax[0] = sns.lineplot(data = df, x = 'remaining_day_to_flight', y= 'price_usd', ax=ax[0])
ax[0].set_title('Ticket Price ($) vs Days Left to Flight', fontsize=20)
plt.xlabel('Day Left to Flight',fontsize=15)
plt.ylabel('Price ($)',fontsize=15)

ax[1] = sns.lineplot(data = df, x = 'remaining_day_to_flight', y= 'price_usd', hue = 'departure_airport')
ax[1].set_title('Ticket Price ($) vs Days Left to Flight', fontsize=20)
plt.xlabel('Day Left to Flight',fontsize=15)
plt.ylabel('Price ($)',fontsize=15)



plt.show()
```

```python
plt.figure(figsize = (16, 8))
sns.lineplot(data = df, x = 'remaining_day_to_flight', y= 'price_usd', hue = 'company')
plt.title('Ticket Price ($) vs Days Left to Flight')
plt.xlabel('Day Left to Flight',fontsize=14)
plt.ylabel('Price ($)',fontsize=14)
plt.show()
```

```python
figsize=(22,20)

ax=sns.regplot(x="price_usd", y="remaining_day_to_flight", data=df,x_bins=25)

ax.set_title('Correlation Between Price ($) and Day Left to Flight', fontsize=20)
ax.set_xlabel('Price ($)',fontsize=14)
ax.set_ylabel('Day Left to Flight',fontsize=14)
```
