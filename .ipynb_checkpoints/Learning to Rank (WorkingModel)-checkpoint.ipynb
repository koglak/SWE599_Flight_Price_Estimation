```python
def get_model_input(X_u, X_m, y, tgt_users):

    merged = pd.merge(X_u, y, on=['remaining_day_to_flight'], how='inner')
    merged = pd.merge(X_m, merged, on=['arrival_city'], how='outer')
    merged = merged.query('remaining_day_to_flight in @tgt_users')
    merged.head()

    merged.fillna(0, inplace=True)
    features_cols = list(merged.drop(columns=['departure_date','company', 'departure_airport_name', 'arrival_city',
       'part_of_day', 'departure_week_day', 'part_of_month', 'is_weekend',
       'price_try', 'remaining_day_to_flight']).columns)

    query_list = merged['remaining_day_to_flight'].value_counts()

    merged = merged.set_index(['remaining_day_to_flight', 'arrival_city'])

    query_list = query_list.sort_index()

    merged.sort_index(inplace=True)

    df_x = merged[features_cols]

    df_y = merged['price_try']
    
    return df_x, df_y, query_list

X_train, y_train, query_list_train = get_model_input(train_X_u, train_X_p, train_y, train_tgt_user)
```

```python
def predict_at_k(data, model, k):
    user_ids = list()
    product_ids = list()
    ranks = list()
    
    for userId, df in data.groupby('remaining_day_to_flight'):
        print(userId)
        pred = model.predict(df.loc[userId])
        print(len(pred))
        productId = np.array(df.reset_index()['arrival_city'])
        topK_index = np.argsort(pred)[::-1][:k]
        product_ids.extend(list(productId[topK_index]))
        user_ids.extend([userId]*len(topK_index))
        ranks.extend(list(range(1, len(topK_index)+1)))
        print()

    results = pd.DataFrame({'remaining_day_to_flight': user_ids, 'arrival_city': product_ids, 'rank': ranks})
    
    return results

predicted = predict_at_k(X_test, model, 5)
```

```python
def coverage(preds,train_X_p):
    test_recs = preds['arrival_city'].nunique()
    train_movies = train_X_p['arrival_city'].nunique()

    return test_recs/train_movies
```
