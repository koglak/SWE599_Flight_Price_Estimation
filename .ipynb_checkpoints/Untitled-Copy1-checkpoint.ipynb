```python
import pandas as pd
import numpy as np
from numpy import mean
from numpy import std

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec

import scipy.stats as ss
from scipy.stats import rankdata
```

```python
df = pd.read_csv(r"C:\Users\kogla\OneDrive\SWE599-Flight-Ticket-Price-Estimation-Project\merged.csv").drop('Unnamed: 0', axis=1)
df.head()
```

```python
df=df.drop(['departure_airport', 'arrival_airport'], axis=1)
```

```python
df = df[['company', 'departure_airport_name', 'arrival_city', 'departure_date', 'departure_time','arrival_time',
         'departure_day', 'departure_month', 'departure_hour', 'part_of_day', 'departure_week_day', 'part_of_month','is_weekend','duration(min)',
        'price_try', 'checked_week_day', 'remaining_day_to_flight' ]]
```

```python
df_cheap=df.copy()
```

```python
df_cheap.head()
```

```python
arrival_cities = df_cheap.arrival_city.unique().tolist()
```

```python
departure_dates = df_cheap.departure_date.unique().tolist()
```

```python
df_rank = pd.DataFrame()

for date in departure_dates:
    for city in arrival_cities:
        filtered= df_cheap.loc[(df_cheap['arrival_city'] == city) & (df_cheap['departure_date'] == date)]
        prices = filtered.price_try.to_numpy()
        filtered['rank']=pd.Series(rankdata(prices,method='dense'))
        df_rank = pd.concat([df_rank, filtered], axis=0).reset_index(drop=True)
```

```python
for city in arrival_cities:
    filtered= df_cheap.loc[(df_cheap['arrival_city'] == city) & (df_cheap['departure_date'] == '2022-11-14')]
    filtered[filtered.price_try==filtered.price_try.min()].remaining_day
```

```python
for date in ['2022-11-14','2022-11-15']:
    for city in arrival_cities:
        filtered= df_cheap.loc[(df_cheap['arrival_city'] == city) & (df_cheap['departure_date'] == date)]
        plt.figure(figsize = (6, 3))
        ax=sns.lineplot(data = filtered, x = 'remaining_day_to_flight', y= 'price_try')
        plt.title('Ticket Price (TL) vs Days Left to Flight')
        plt.xlabel('Days Left to Flight for {} / {}'.format(city,date),fontsize=14)
        plt.ylabel('Price (TL)',fontsize=14)
        ax.set(xlim=(0, None))
        plt.show()
```
