```python
# importing libraries
import pandas as pd
import glob
import os

# merging the files
joined_files = os.path.join(r"C:\Users\kogla\OneDrive\Masaüstü\SWE599\dataset", "*.csv")

# A list of all joined files is returned
joined_list = glob.glob(joined_files)

# Finally, the files are joined
df = pd.concat(map(pd.read_csv, joined_list), ignore_index=True).drop('Unnamed: 0', axis=1)

df.head()
```

```python
df[['departure_day','departure_month', 'departure_year_week']] = df[['A','B']].apply(pd.to_datetime) #if conversion required
df['C'] = (df['B'] - df['A']).dt.days
```

```python
df[['V','allele']] = df['departure_date'].str.split(',',expand=True)
```
